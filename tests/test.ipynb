```python
import matplotlib.pyplot as plt


#     
def loss_function(x):
    return (x - 5) ** 2


def derivative_loss_function(x):
    return 2 * (x - 5)


#   
learning_rate = 0.1
initial_guess = 0

#      
x_updates = [initial_guess]
loss_updates = [loss_function(initial_guess)]

#   
x = initial_guess
for i in range(20):
    grad = derivative_loss_function(x)
    x = x - learning_rate * grad
    x_updates.append(x)
    loss_updates.append(loss_function(x))

# 
plt.figure(figsize=(12, 5))

#    x
plt.subplot(1, 2, 1)
plt.plot(x_updates, "o-")
plt.title("  x")
plt.xlabel("")
plt.ylabel(" x")

#     
plt.subplot(1, 2, 2)
plt.plot(loss_updates, "r-s")
plt.title("  ")
plt.xlabel("")
plt.ylabel("  ")

plt.tight_layout()
plt.show()
```
